```python
import pandas as pd
import numpy as np
```

```python
years = np.arange(2003,2024)
d = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct': 10, 'Nov':11, 'Dec':12}
```

```python
xls = pd.ExcelFile('CPIHistoricalForecast.xlsx')

for year in years:
  locals()["df_"+str(year)] = pd.read_excel(xls, f'{year}', header=2, index_col=0)
  locals()["df_"+str(year)] = locals()["df_"+str(year)][locals()["df_"+str(year)].index.notnull()][0:-6]

  locals()["df_"+str(year)] = locals()["df_"+str(year)].stack().reset_index()

  for i in range(0,int(locals()["df_"+str(year)].shape[0]/3)):
    locals()["df_"+str(year)].at[i*3+1, 'lower'] = locals()["df_"+str(year)].at[i*3, 0]
    locals()["df_"+str(year)].at[i*3+1, 'mid'] = locals()["df_"+str(year)].at[i*3+1, 0]
    locals()["df_"+str(year)].at[i*3+1, 'upper'] = locals()["df_"+str(year)].at[i*3+2,0]
    locals()["df_"+str(year)].drop([i*3,i*3+2], inplace=True)


    new = locals()["df_"+str(year)].at[i*3+1, "level_1"].split("-")
    locals()["df_"+str(year)].at[i*3+1, "curr_month"] = new[0]
    locals()["df_"+str(year)].at[i*3+1, "curr_year"] = 2000 + int(new[1])

  locals()["df_"+str(year)].curr_month = locals()["df_"+str(year)].curr_month.map(d)
  locals()["df_"+str(year)].reset_index(inplace=True, drop=True)
  locals()["df_"+str(year)].drop(['level_1', 0], 1, inplace=True)
  locals()["df_"+str(year)].rename(columns={locals()["df_"+str(year)].columns[0]: 'cpi_item'}, inplace=True)
  locals()["df_"+str(year)]['pred_year'] = year
  locals()["df_"+str(year)] = locals()["df_"+str(year)][['cpi_item', 'curr_month', 'curr_year', 'lower', 'mid', 'upper', 'pred_year']]

  locals()["df_"+str(year)].to_csv(f"cpi_historical_forecast_{year}.csv", index=False)
```
