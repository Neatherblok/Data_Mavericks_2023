```python
# Import necessary library
import pandas as pd
```

```python
# Read PPI changes excel and convert to Pandas dataframe
df = pd.DataFrame(pd.read_excel(f"Producer Price Index change.xlsx", header=1, index_col = 0))
```

```python
# To remove unnecessary rows and columns
df = df.iloc[:-5, :-6]

# To remove index row with index number NaN
df = df[df.index.notnull()][0:-3]
df
```

```python
# Flatten year columns to row values
df = df.stack().reset_index()
```

```python
# Rename flattened columns to year and PPI annual change in percentage.
df.rename(columns={df.columns[0]: 'ppi_item', df.columns[1]: 'year', df.columns[2]: 'ppi_annual_change_in_perc'}, inplace=True)
```

```python
# Sort dataframe by year value and in order of original appearance in Excel file
df = df.reset_index().sort_values(by=['year', 'index']).iloc[:,1:]
```

```python
# Save in new CSV file
df.to_csv(f"ppi_change.csv", index=False)
```
