```python
import pandas as pd
```

```python
#seperately look at pork
article = "Beef chunks"
d = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct': 10, 'Nov':11, 'Dec':12}
```

```python
df = pd.DataFrame(pd.read_excel(f"{article}.xlsx", header=9, index_col = 0))
```

```python
## For pork Excel file

# df['month'] = df.index.month
# df['year'] = df.index.year
```

```python
df = df.stack().reset_index()
```

```python
df.rename(columns={df.columns[0]: 'year', df.columns[1]: 'month', df.columns[2]: 'us_city_average_retail_price'}, inplace=True)
df['product_name'] = article
df = df[['product_name', 'year', 'month','us_city_average_retail_price']]
```

```python
df.month = df.month.map(d)
```

```python
df = df.sort_values(by=['year', 'month'])
```

```python
df.to_csv(f"{article.lower().replace(' ', '_')}.csv", index=False)
```
