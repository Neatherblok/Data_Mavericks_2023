```python
# Merge the data
merged_df = pd.merge(df, inflation_df, on='year')

# Define the categories and their respective products
categories = {
    'Animal Products': ['Eggs', 'Beef steaks', 'Beef chunks', 'Chicken', 'Pork'],
    'Dairy Products': ['American Cheese', 'Cheddar cheese', 'Milk'],
    'Fruits': ['Bananas', 'Oranges', 'Strawberries', 'Tomatoes'],
    'Grains': ['Rice', 'Flour', 'Bread']
}

# Define the color map
cmap = plt.get_cmap('tab10')

# Loop through each category and plot the data
for i, category in enumerate(categories):
    # Get the products for this category
    category_products = categories[category]

    # Create a groupby object for this category
    grouped_df = merged_df.groupby(['category', 'product_name'])

    # Create a new figure
    fig, axs = plt.subplots(figsize=(10, 6))
    axs.set_xlabel('Inflation Rate (%)')
    axs.set_ylabel('Retail Price (USD)')

    # Loop through each product in this category and plot the data
    for j, product in enumerate(category_products):
        # Get the data for this product
        product_df = grouped_df.get_group((category, product))

        # Get a color from the color map based on the index of the product in the group
        color = cmap(grouped_df.groups[(category, product)].tolist()[0] % cmap.N)

        # Plot the scatter plot for this product
        axs.scatter(product_df['inflation_rate_pct'], product_df['us_city_average_retail_price'], color=color, label=product)

        # Calculate the correlation coefficient
        corr_coef = product_df[['inflation_rate_pct', 'us_city_average_retail_price']].corr(method='pearson')['us_city_average_retail_price'][0]

        # Add the correlation coefficient to the plot title
        axs.set_title(f"{category} ({product}): Correlation Coefficient = {corr_coef:.2f}")

    # Add a legend to the plot
    axs.legend()

    # Save the plot
    plt.savefig(f"{category}.png")
```
